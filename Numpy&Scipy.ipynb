```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(-2*np.pi,2*np.pi,1/1000)
```

```python
y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)
```

```python
plt.plot(x,y_sin,linestyle='dashed')
plt.plot(x,y_cos,linestyle='dashdot')
plt.plot(x,y_tan,'.',markersize=.5)

plt.ylim([-1.2,1.2])
plt.legend(('sin','cos','tan'))
```
